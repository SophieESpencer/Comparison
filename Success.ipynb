```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
df= pd.read_excel('pivot_test.xlsx')
```

```python
df.info()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
df_no = df[df['ASBH02A']=='No']
```

```python
df_no
```

```python

df_no_melted = df_no.melt(id_vars='Country', var_name='Metric', value_name='Value')
df_no_melted
```

```python
# Set the figure size
plt.figure(figsize=(12, 8))

# Create the bar plot
sns.barplot(x='Country', y='Value', hue ='Metric', data=df_no_melted)

# Move the legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Value')
plt.title('Bar Plot of Informational Purpose Average')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
df_yes= df[df['ASBH02A']=='Yes']
df_yes = df_yes.drop('ASBH02A', axis=1)
```

```python
df_yes_melted = df_yes.melt(id_vars='Country', var_name='Metric', value_name='Value')
df_yes_melted
```

```python
# Set the figure size
plt.figure(figsize=(12, 8))

# Create the bar plot
sns.barplot(x='Country', y='Value', hue ='Metric', data=df_yes_melted)

# Move the legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Value')
plt.title('Bar Plot of Informational Purpose Average')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
bar_width = 0.8

# Create figure and axis
fig, ax = plt.subplots()

# Plot larger bars
ax.bar(df_yes_melted['Country'], df_yes_melted['Value'], bar_width, label='yes', alpha=0.6)

# Plot smaller bars on top of the larger bars
ax.bar(df_no_melted['Country'], df_no_melted['Value'], bar_width, label='no', alpha=1.0)

# Add labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Value')
ax.set_title('Overlapping Bar Chart')


# Set x-ticks and x-tick labels
ax.set_xticks(range(len(df_yes_melted['Metric'])))
ax.set_xticklabels(df_yes_melted['Metric'])

# Rotate x-axis labels
plt.xticks(rotation=45)

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Plot larger bars
ax.bar(df_yes_melted.index, df_yes_melted['Value'], bar_width, label='yes', alpha=0.6)

# Plot smaller bars on top of the larger bars
ax.bar(df_no_melted.index, df_no_melted['Value'], bar_width, label='no', alpha=1.0)

# Add labels and title
ax.set_xlabel('Country-Metric')
ax.set_ylabel('Value')
ax.set_title('Overlapping Bar Chart')

# Set x-ticks and x-tick labels
ax.set_xticks(df_yes_melted.index)
ax.set_xticklabels([f"{country}-{metric}" for country, metric in zip(df_yes_melted['Country'], df_yes_melted['Metric'])], rotation=45, ha='right')

# Add legend
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Define bar width
bar_width = 0.4
n_metrics = df_no_melted['Metric'].nunique()

# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Get unique countries and metrics
countries = df_no_melted['Country'].unique()
metrics = df_no_melted['Metric'].unique()

# Create x indices for each group of bars
indices = np.arange(len(countries)) * (n_metrics + 1)

# Plot each group of bars for 'yes' and 'no'
for i, metric in enumerate(metrics):
    yes_values = df_yes_melted[df_yes_melted['Metric'] == metric].set_index('Country').reindex(countries)['Value']
    no_values = df_no_melted[df_no_melted['Metric'] == metric].set_index('Country').reindex(countries)['Value']
    
    ax.bar(indices + i * bar_width, yes_values, bar_width, label=f'yes ({metric})', alpha=0.6)
    ax.bar(indices + i * bar_width, no_values, bar_width, label=f'no ({metric})', alpha=1.0)

# Add labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Value')
ax.set_title('Overlapping Bar Chart by Metric and Country')

# Set x-ticks and x-tick labels
ax.set_xticks(indices + bar_width * (n_metrics - 1) / 2)
ax.set_xticklabels(countries, rotation=45, ha='right')

# Add legend
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Define bar width
bar_width = 0.8
n_metrics = df_no_melted['Metric'].nunique()

# Colors for each metric
colors = {
    'informational_purpose_avg': 'red',
    'interpreting_process_avg': 'blue',
    'literary_purpose_avg': 'green',
    'reading_avg': 'orange',
    'straightforward_process_avg': 'purple'
}

# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Get unique countries and metrics
countries = df_no_melted['Country'].unique()
metrics = df_no_melted['Metric'].unique()

# Create x indices for each group of bars
indices = np.arange(len(countries)) * (n_metrics + 1)

# Plot each group of bars for 'yes' and 'no'
for i, metric in enumerate(metrics):
    yes_values = df_yes_melted[df_yes_melted['Metric'] == metric].set_index('Country').reindex(countries)['Value']
    no_values = df_no_melted[df_no_melted['Metric'] == metric].set_index('Country').reindex(countries)['Value']
    
    color = colors[metric]
    ax.bar(indices + i * bar_width, yes_values, bar_width, label=f'yes ({metric})', color=color, alpha=0.6)
    ax.bar(indices + i * bar_width, no_values, bar_width, label=f'no ({metric})', color=color, alpha=1.0)

# Add labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Value')
ax.set_title('Overlapping Bar Chart by Metric and Country')

# Set x-ticks and x-tick labels
ax.set_xticks(indices + bar_width * (n_metrics - 1) / 2)
ax.set_xticklabels(countries, rotation=45, ha='right')

# Add legend
ax.legend()

# Move the legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Define bar width
bar_width = 1
n_metrics = df_no_melted['Metric'].nunique()

# Colors for each metric
colors = {
    'informational_purpose_avg': 'red',
    'interpreting_process_avg': 'blue',
    'literary_purpose_avg': 'green',
    'reading_avg': 'orange',
    'straightforward_process_avg': 'purple'
}

# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Get unique countries and metrics
countries = df_no_melted['Country'].unique()
metrics = df_no_melted['Metric'].unique()

# Create x indices for each group of bars
indices = np.arange(len(countries)) * (n_metrics + 1)

# Plot each group of bars for 'yes' and 'no'
for i, metric in enumerate(metrics):
    yes_values = df_yes_melted[df_yes_melted['Metric'] == metric].set_index('Country').reindex(countries)['Value']
    no_values = df_no_melted[df_no_melted['Metric'] == metric].set_index('Country').reindex(countries)['Value']
    
    color = colors[metric]
    ax.bar(indices + i * bar_width, yes_values, bar_width, color=color, alpha=0.6)
    ax.bar(indices + i * bar_width, no_values, bar_width, color=color, alpha=1.0)

# Add labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Value')
ax.set_title('Overlapping Bar Chart by Metric and Country')

# Set x-ticks and x-tick labels
ax.set_xticks(indices + bar_width * (n_metrics - 1) / 2)
ax.set_xticklabels(countries, rotation=45, ha='right')

# Create custom legend
from matplotlib.patches import Patch
legend_handles = [Patch(color=color, label=metric) for metric, color in colors.items()]
ax.legend(handles=legend_handles, title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')


# Show plot
plt.tight_layout()
plt.show()
```
