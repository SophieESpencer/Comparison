```python
assessment_score_columns = ['ASRREA01', 'ASRREA02', 'ASRREA03', 'ASRREA04', 'ASRREA05', 'ASRLIT01', 'ASRLIT02', 'ASRLIT03', 'ASRLIT04', 'ASRLIT05', 'ASRINF01', 'ASRINF02', 'ASRINF03', 'ASRINF04', 'ASRINF05', 'ASRIIE01', 'ASRIIE02', 'ASRIIE03', 'ASRIIE04', 'ASRIIE05', 'ASRRSI01', 'ASRRSI02', 'ASRRSI03', 'ASRRSI04', 'ASRRSI05']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pivot_df_multiplexreadingscore = df.pivot_table(values=assessment_score_columns, index=['Country','ASBH02A'], aggfunc='mean')
pivot_df_multiplexreadingscore
```

```python
ax = pivot_df_multiplexreadingscore.plot(kind='bar',figsize=(15,10),width = 0.8)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
# Move the legend outside the plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Legend')
plt.show()
```

```python
pivot_df_multiplexreadingscore_unstacked = pivot_df_multiplexreadingscore.unstack(level=1)
ax2 = pivot_df_multiplexreadingscore_unstacked.plot(kind='bar', figsize=(15,10), width=0.8)
plt.title('Mean Reading Scores by Country and ASBH02A')
plt.xlabel('Country')
plt.ylabel('Mean Score')
plt.legend(title='ASBH02A')
# Move the legend outside the plot
ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Legend')
plt.show()
```

Let's chang colours so it's a bit clearer.

```python
# Define color mapping for ASBH02A values
color_mapping = {
    'No': 'darkblue',
    'Yes': 'lightblue'
}

# Function to get colors based on column name
def get_colors(columns, color_mapping):
    return [color_mapping[col] for col in columns]

# Get colors for the bars
bar_colors = get_colors(pivot_df_multiplexreadingscore_unstacked.columns.get_level_values(1), color_mapping)

# Plot the unstacked DataFrame
ax4 = pivot_df_multiplexreadingscore_unstacked.plot(kind='bar', figsize=(15,10), width=0.8, color=bar_colors)
plt.title('Mean Reading Scores by Country and ASBH02A')
plt.xlabel('Country')
plt.ylabel('Mean Score')

# Move the legend outside the plot
ax4.legend(title='ASBH02A', loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
# Step 2: Calculate the mean score for each country
country_mean_scores = pivot_df_multiplexreadingscore.groupby(level='Country').mean()

# If you want to sum the mean scores across all assessment columns (if there are multiple), you can do:
country_mean_scores['MeanTotal'] = country_mean_scores.mean(axis=1)

# Sort countries by total values
sorted_countries_means = country_mean_scores.sort_values(by='MeanTotal', ascending=False).index

# Create a custom sorter for ASBH02A
asbh02a_sorter = ['Yes', 'No']

# Get the current index as a DataFrame for sorting
index_df_reading = pivot_df_multiplexreadingscore.index.to_frame()

# Rename the columns to avoid conflict
index_df_reading = index_df_reading.rename(columns={'Country': 'Country_', 'ASBH02A': 'ASBH02A_'})

# Sort the DataFrame first by Country using the sorted_countries and then by ASBH02A using the custom sorter
index_df_reading['Country_'] = pd.Categorical(index_df_reading['Country_'], categories=sorted_countries_means, ordered=True)
index_df_reading['ASBH02A_'] = pd.Categorical(index_df_reading['ASBH02A_'], categories=asbh02a_sorter, ordered=True)
index_df_reading = index_df_reading.sort_values(by=['Country_', 'ASBH02A_'])

# Reindex the pivot table using the sorted index
sorted_pivot_df_reading = pivot_df_multiplexreadingscore.loc[index_df_reading.index]

# Plot the heat map
plt.figure(figsize=(10, 8))
sns.heatmap(sorted_pivot_df_reading, annot=False, cmap='YlGnBu')
plt.title('Reading Scores Heat Map Sorted by Country Mean and ASBH02A')
plt.show()
```

```python
df["reading_avg"] = df[reading].mean(axis=1)
```

```python
# Define the function
def calculate_average(df, columns, new_column_name):
    """
    Calculates the mean of the specified columns and adds a new column with the result.
    
    Parameters:
    df (pd.DataFrame): The input DataFrame.
    columns (list): The list of column names to calculate the mean.
    new_column_name (str): The name of the new column to store the averages.
    
    Returns:
    pd.DataFrame: The DataFrame with the new column added.
    """
    df[new_column_name] = df[columns].mean(axis=1)
    return df

# Calculate the mean of the reading columns
df = calculate_average(df, reading, 'reading_avg')
df = calculate_average(df, literary_purpose, 'literary_purpose_avg')
df = calculate_average(df, informational_purpose, 'informational_purpose_avg')
df = calculate_average(df, interpreting_process, 'interpreting_process_avg')
df = calculate_average(df, straightforward_process, 'straightforward_process')
```
